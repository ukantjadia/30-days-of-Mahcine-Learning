```python
import pandas as pd
import numpy as np
# https://stevealbertwong.github.io/2016/12/23/SVD/
col = ['hp','dell','macbook','thinbook','notebook','acer']
x =  np.array([
	[1,1,0.5,3,4,5],
	[2,2,1,4,1,4],
	[3,3,1.5,2,5,1],
	[4,4,2,3,4,2],
	[1,3,2.2,1,3,4],
	[5,5,3,4.1,4,3],
	[2,3,3.5,2,5,3],
	[2,1,4,5,4,5]])
data=pd.DataFrame(data=x,columns=col)
data
```

```python
from numpy import array
from scipy.linalg import svd

# https://setosa.io/ev/principal-component-analysis/
U,Sigma ,V_transpose = svd(data,full_matrices=False)
print(f'The covariance of matrix is \n {np.dot(data.T,data)}')
print(f'\nThe U left sigular matrix\n {U}')
print(f'\nThe Sigma diagonal/eigenvalues matrics\n {Sigma}')
print(f'\nThe V the right singular 
\n {V_transpose}')
```
