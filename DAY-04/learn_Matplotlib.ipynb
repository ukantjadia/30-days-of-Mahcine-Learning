Below code can be done in two ways 
1. Object-oriented style
2. pyploy-style

---



```python
# 1 : Object-oriented style
x = np.linspace(0,6,10) # sample data linspace-- create the data with linear space

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,x,label='Linear')
ax.plot(x,x**2,label='Quadratic')
ax.plot(x,x**3,label='Cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend();
```

```python
# 2 : pypot-style
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
```

## Styling Artist
- this can be use to add manually color, linewidth, linestyle on plot

### Some differetn linestyle
- `o` as circle
- `d` as diamond
- `v` as triangle
- `s` as square

### Color

## Object oriented way to plot

```python
f = plt.figure() # creating a object of class figure
a1 = f.add_axes([0,0,1,1]) # using that obj to create various plot
a2 = f.add_axes([1.1,0,1,1])
a3 = f.add_axes([0.1,0.1,.5,.5])
a1.plot(d1,'o',label='d1') # passing data to the plot using plot name
a2.plot(d3,d4,'d',label='d3 and d4')
a3.plot(d4,'s')
a3.set_title('test') # following all setting some parameter for better plot
a3.set_xlabel('value')
a3.set_ylabel('num')
a1.legend()
```

## Setting the ticks manually 


```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(d1, d2)
axs[0].set_title('Automatic ticks')

axs[1].plot(d1, d2)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])  # note that we don't need to specify labels
axs[1].set_title('Manual ticks');
```
